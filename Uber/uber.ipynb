```python
import datetime as dt
df['Request timestamp']=df['Request timestamp'].replace('/', '-', regex=True)
df['Request timestamp']=pd.to_datetime(df['Request timestamp'],format='mixed')
df['Drop timestamp']=pd.to_datetime(df['Drop timestamp'],format='mixed')

df['Request Date'] =pd.to_datetime(df['Request timestamp']).dt.normalize()
df['Request Time'] = pd.to_datetime(df['Request timestamp']).dt.time
#df['Request Time'] = pd.to_datetime(df['Request Time'], format='%H:%M:%S')
df['Drop Date'] = pd.to_datetime(df['Drop timestamp']).dt.normalize()
df['Drop Time'] = pd.to_datetime(df['Drop timestamp']).dt.time
#df['Drop Time'] = pd.to_datetime(df['Drop Time'], format='%H:%M:%S')
#df=df.drop(['Drop timestamp', 'Request timestamp'], axis=1)


```

```python
df['duration']=pd.to_datetime(df['Request Time'], format='%H:%M:%S') - pd.to_datetime(df['Request Time'], format='%H:%M:%S').dt.normalize()
```

```python
df.info()
```

##### Maximum duration ####

```python
df.loc[df['duration'] == df['duration'].max()]
```
